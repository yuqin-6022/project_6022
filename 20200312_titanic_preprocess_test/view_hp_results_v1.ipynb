```python
def plot_history(histories):
    line_types = ('-', '-o')
    
    histories_index = tuple(range(len(histories)))
    colors = ['r', 'g', 'b']
    
    plt.figure()
    for i, history, color in zip(histories_index, histories, colors):
        fmt = '%s%s' % (color, line_types[0])
        val_fmt = '%s%s' % (color, line_types[1])
        loss, val_loss = history['loss'], history['val_loss']
        plt.plot(loss, fmt, label=str(i))
        plt.plot(val_loss, val_fmt, label=str(i))
    plt.legend()
    plt.title('loss')
    
    plt.figure()
    for i, history, color in zip(histories_index, histories, colors):
        fmt = '%s%s' % (color, line_types[0])
        val_fmt = '%s%s' % (color, line_types[1])
        accuracy, val_accuracy = history['accuracy'], history['val_accuracy']
        plt.plot(accuracy, fmt, label=str(i))
        plt.plot(val_accuracy, val_fmt, label=str(i))
    plt.legend()
    plt.title('accuracy')
```

```python
# 
ENCODED_PATH = os.path.join(CUR_PATH, FILE_NAME, ENCODED_RESULE)
with open(ENCODED_PATH, 'r') as f:
    results = json.load(f)

encoded_histories = tuple(json.loads(model_test_result['history'].replace('\'', '"')) for model_test_result in results['model_test_results'])
plot_history(encoded_histories)
```
