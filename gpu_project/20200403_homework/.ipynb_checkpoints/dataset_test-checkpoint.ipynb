```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.preprocessing import StandardScaler
from sklearn.utils import check_random_state
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./dataset.csv')
```

```python
df.info()
df.describe()
```

```python
df.head()
```

```python
random_state = check_random_state(0)
permutation = random_state.permutation(df.shape[0])
df = df.iloc[permutation, :]

x = df.iloc[:, list(range(11))].copy().values
y_dloc = df['dloc'].copy().values
y_ed = df['ED'].copy().values
y_ed = df['overload_loc'].copy().values
```

```python
TEST_SIZE = 0.1

x_train, x_test, y_train, y_test = train_test_split(x, y_dloc, test_size=TEST_SIZE)
print(x_train.shape)
print(x_test.shape)
```

```python
clf = LogisticRegressionCV(penalty='l2', solver='sag', tol=0.001, multi_class='multinomial', max_iter=10000)
```

```python
clf.fit(x_train, y_train)
```

```python
score = clf.score(x_test, y_test)
print(score)
```
